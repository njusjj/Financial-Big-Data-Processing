```python
import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties
import seaborn as sns


font = FontProperties(size=20)
plt.figure(figsize=(15, 10))
plt.bar([1],[0.942818], width=0.4, label='Logistics')
plt.bar([2],[0.943163], width=0.4,label='SVM')
plt.bar([3],[0.933072], width=0.4,label='DecisionTree')
plt.bar([4],[0.942307], width=0.4,label='RandomForest')
plt.xticks([1,2,3,4],['Logistics','SVM','DecisionTree','RandomForest'],FontProperties=font)
plt.ylim([0.93,0.95])
plt.grid(color='gray', linestyle='-', linewidth=.5)  
plt.legend(ncol=4, loc=9,fontsize=20)

plt.xlabel('Methods',FontProperties=font)
plt.ylabel('Accuracy',FontProperties=font)

plt.title(u'Comparison', FontProperties=font)

plt.show()
plt.savefig('/Users/shijianjun/学习资料/金融大数据处理/作业/实验3/comparison')
```
